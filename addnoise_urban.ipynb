```python
import soundata
import numpy as np
import os
from scipy.io.wavfile import write
from tqdm import tqdm
```

```python
# Load the dataset
import os
dataset_path = os.path.expanduser("/data/users/tqiu5/CleanUNet/urbansound8k")
urbansound8k = soundata.initialize('urbansound8k',data_home=dataset_path)
if  os.path.exists(dataset_path):
    print("Dataset downloaded, continue")
else:
    urbansound8k.download()
```

```python
# Set output path for noisy clips
noisy_output_dir = "/data/users/tqiu5/CleanUNet/urbansound8k/audio/fold1_noisy"
clean_output_dir = "/data/users/tqiu5/CleanUNet/dns/datasets/test_set/synthetic/no_reverb/urban_set/clean"
os.makedirs(noisy_output_dir, exist_ok=True)

# Noise parameters
noise_level = 0.1  # 10% of max amplitude

# Process all clips
i = 0
for clip_id in tqdm(urbansound8k.clip_ids):
    clip = urbansound8k.clip(clip_id)

    if clip.fold != 1:
        continue

    audio, sr = clip.audio
    
    # Generate and add Gaussian noise
    noise = np.random.normal(0, noise_level * np.max(np.abs(audio)), len(audio))
    noisy_audio = audio + noise
    
    # Clip to valid range [-1, 1] to avoid distortion
    noisy_audio = np.clip(noisy_audio, -1.0, 1.0)

    # Save the noisy audio
    noisy_output_path = os.path.join(noisy_output_dir, f"noisy_fileid_{clip_id}.wav")
    clean_output_path = os.path.join(clean_output_dir, f"clean_fileid_{clip_id}.wav")
    write(noisy_output_path, sr, (noisy_audio * 32767).astype(np.int16))
    write(clean_output_path, sr, (audio * 32767).astype(np.int16))

    i = i + 1
```

```python
import os
import torchaudio
from torchaudio.transforms import Resample

input_folder = '/data/users/tqiu5/CleanUNet/exp/DNS-large-high/speech_urban/150k'
output_folder = '/data/users/tqiu5/CleanUNet/exp/DNS-large-high/speech_urban/150k'
target_sr = 16000

os.makedirs(output_folder, exist_ok=True)

for fname in os.listdir(input_folder):
    if fname.endswith('.wav'):
        in_path = os.path.join(input_folder, fname)
        out_path = os.path.join(output_folder, fname)

        waveform, sr = torchaudio.load(in_path)

        if sr != target_sr:
            resampler = Resample(orig_freq=sr, new_freq=target_sr)
            waveform = resampler(waveform)

        torchaudio.save(out_path, waveform, target_sr)
        print(f"Saved {out_path}")
```
